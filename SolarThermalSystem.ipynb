# Solar Thermal System Simulation & Validation

This notebook simulates a solar thermal collector and storage tank system using the standard Hottel-Whillier model and tank energy balance ODEs.  
It visualizes system temperatures, solar input, energy balance, and collector efficiency for engineering validation.



Water flows in a closed loop:  
Tank → Collector → Pump → Tank.

```
   Sunlight
      ↓
      │              (Hot fluid)
      [Solar Panel]──────────────[ Pump ]───→ [Storage Tank]
      ↑                                                    ↓
      │             (Cooled fluid)                         │
      └────────────────────────────────────────────────────┘
```
